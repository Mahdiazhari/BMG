```python
import pandas as pd
import numpy as np
import re
```

```python
dataset = pd.read_csv('temp_jogja.csv', delimiter=',', header=0)
dataset.head()
```

```python
dataset['ad_link']=dataset['ad_link'].str.lower()
dataset['title']=dataset['title'].str.lower()
dataset['make']=dataset['make'].str.lower()
dataset['model']=dataset['model'].str.lower()
dataset['machine_capacity']=dataset['machine_capacity'].str.lower()
dataset['variant']=dataset['variant'].str.lower()
dataset['transmission']=dataset['transmission'].str.lower()
dataset['fuel']=dataset['fuel'].str.lower()
dataset['description']=dataset['description'].str.lower()
dataset.head()
```

```python
dataset=dataset.fillna(' ')
dataset = dataset[~dataset['machine_capacity'].str.contains('km',na=False)] #remove machine capacity with "km"
dataset.head()
```

```python
#take machine_capacity column
check_machine = dataset['machine_capacity']
initialize_list = [] #initialize columns for adding the final capacity values
for string in check_machine:
    cc= re.findall('\d+',string)
    if len(cc)>0:
        capacity=int(cc[0])/1000
        initialize_list.append(float(capacity))
    else:
        initialize_list.append("")
dataset['capacity'] = initialize_list
#dataset['capacity']

check_variant = dataset['variant'].values
capacities = dataset['capacity'].values


for number in range(0,len(capacities)):
    if capacities[number] == '':    
        get_float = re.findall("\d+\.\d+", check_variant[number])
        if len(get_float)>0:
            capacities[number] = get_float[0]
        else:
            capacities[number] = ''



dataset['capacity']=capacities
```

```python
#Fixing/Matching the Years
regex =  "\d{4}" #set the regex to find 4 digit numbers
list_year = dataset['year'].tolist()
list_title = dataset['title'].tolist()

type(list_title)
type(list_year)
print(list_title[2])

for i in range(0,len(list_year)):
    match= re.findall(regex, list_title[i])
    if len(match)>0:
        year_title = int(match[0])
        if year_title > list_year[i]:
            list_year[i] = year_title
            
dataset['year'] = list_year
dataset.head()

    
    
```
